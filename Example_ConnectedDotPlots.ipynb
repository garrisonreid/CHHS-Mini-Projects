## **Connected Dot Plots**
In this assignment, we'll create two connected dot plot charts. The data should come from two seperate datasets: Sheets you create, online files, or APIs.

**Documentation:** https://python-graph-gallery.com/184-lollipop-plot-with-2-groups/

**Example charts:** https://www.storytellingwithdata.com/blog/2018/8/21/august-swdchallenge-recap-the-dot-plot

### Import common graphing and analysis libraries

```
# Import your libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## **Example with comments**

```
# YOU WILL NOT NEED TO DO THIS SINCE YOUR DATA WILL COME FROM ANOTHER SOURCE
## Create a dataframe
value1=np.random.uniform(size=20)
value2=value1+np.random.uniform(size=20)/4
df = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'value1':value1 , 'value2':value2 })
df.head()
```

```
# Reorder it following the values of the first value:
ordered_df = df.sort_values(by='value1')
my_range=range(1,len(df.index)+1)
 
# This changes the size of the graphic
plt.figure(figsize=(20,10))

# This creates soft lines for the eye to follow
plt.hlines(y=my_range, xmin=ordered_df['value1'].min(), xmax=ordered_df['value2'].max(), color='grey', alpha=0.15, linestyles="dotted")

# This adds the line that connects the points
plt.hlines(y=my_range, xmin=ordered_df['value1'], xmax=ordered_df['value2'], color='grey', alpha=0.7, linestyles="dotted")

# This adds the scatter plot points for the initial column of data
plt.scatter(ordered_df['value1'], my_range, color='skyblue', alpha=1, label='value1', s=50, zorder=50)

# This adds the scatter plot points for the second column of data
plt.scatter(ordered_df['value2'], my_range, color='green', alpha=0.4 , label='value2', s=50, zorder=50)

# This adds a legend to the graph
plt.legend()
 
# Adds, ticks, title, and axis labels
plt.yticks(my_range, ordered_df['group'])
plt.title("Comparison of the value 1 and the value 2", loc='left', fontdict = {'fontsize':20, 'fontweight':'semibold'})
plt.xlabel('Value of the variables')
plt.ylabel('Group')

# Displays chart
plt.show()
```
