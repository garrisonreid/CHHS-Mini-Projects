```
restaurant = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSgx1QP1wl_rXzxeouCpztMoPPDFQCxXF0Z5XTWphsWU-pg09oec0jXteIfAQHjnR-H3ZlouVyVG5f0/pub?output=csv")
```

```
locations = restaurant[['Lat','Long' ]]
locationlist = locations.values.tolist()
colors = []
for _ , point in restaurant["Final Score"].iteritems():
  if point >= 100:
    colors.append('pink')
  elif point >= 95:
      colors.append('lightgreen')
  elif point >= 90:
     colors.append('green')
  elif point >= 85:
    colors.append("darkgreen")
  elif point >= 80:
    colors.append("orange")
  else:
    colors.append("red")
```

```
map = folium.Map([35.913237,	-79.079676], zoom_start = 15)
for point in range(0, len(locationlist)):
  folium.Marker(locationlist[point], popup = str(restaurant['Name'][point]) + '\nGrade: '  + str(restaurant['Final Score'][point]), icon = folium.Icon(color = colors[point])).add_to(map)
map
```
